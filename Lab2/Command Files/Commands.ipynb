```python
import pandas as pd
import numpy as np

data = pd.read_csv("C:/Users/user/Desktop/BD/Lab2/Orginal Data/billboard.csv") # wczytanie danych
df = pd.DataFrame(data)

df
```

```python
df_new = df.melt(id_vars=["year","artist.inverted", "track", "time", "genre", "date.entered", "date.peaked"], 
        var_name="week", 
        value_name="no.")

# zamiana kolumn tygodni w jedną kolumnę oraz określenie pozycji w drugiej

df_new
```

```python
print(np.unique(df_new[['year']].values)) # wszytkie piosenki w 2000 roku, dlatego usuwam daną wartość
print(np.unique(df_new[['genre']].values))
```

```python
df_new_2 = df_new.dropna(subset=['no.']) # usunięcie wartości NaN w no.

df_new_2
```

```python
df_new_3 = df_new_2.sort_values(by=['track', 'artist.inverted']) # sortowanie
df_new_4 = df_new_3.drop(columns=['year']) # usuniecie tych samych danych
df_new_5 = (df_new_4.reset_index()).drop(columns=['index']) # nowe indeksy


df_new_5
```

```python
df_new_5.to_csv("C:/Users/user/Desktop/BD/Lab2/Analysis Data/billboard_new.csv") # zapis nowego pliku
```
