## Zadanie 3
Zdefiniuj poniższą funkcję i sporządź jej wykres dla argumentów z danego przedziału:

$f(x)=x^2+5$

$x>-1$  oraz  $x<1$

$x>-6$  oraz  $x<6$

$x>0$  oraz  $x<5$

Wspierając się dokumentacją Matplotlib. Dodaj do wykresu etykiety osi, tytuły wykresów i legendy.

Wskazówki: Deklaracja funkcji w Python.

W Pythonie bloki kodu (w tym również funkcje) są wyróżniane za pomocą wcięć. Do wizualizacji wyników w Pythonie używa się pakietu Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun(x): return x*x+5

x1 = np.linspace(-1,1,1000)
x2 = np.linspace(-6,6,1000)
x3 = np.linspace(0,5,1000)

plt.figure(figsize=(18,5))

ax1 = plt.subplot(131)
plt.plot(x1, fun(x1))
plt.legend(['Od -1 do 1'],loc='lower right')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

ax2 = plt.subplot(132)
plt.plot(x2, fun(x2))
plt.legend(['Od -6 do 6'],loc='lower right')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

ax3 = plt.subplot(133)
plt.plot(x3, fun(x3))
plt.legend(['Od 0 do 5'],loc='lower right')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()

plt.suptitle('Wykresy funkcji w zależności od przedziału argumentów',fontsize=16)
plt.show()
```
